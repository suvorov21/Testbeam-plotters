# Multiplicity

```python
%jsroot on
ROOT.gStyle.SetOptTitle(0)
c.cd()

f = 9
hs = ROOT.THStack()
h_a = []
h_sup = []

for num, f in enumerate(file_list):
    print('working on file', num+1, 'of', len(file_list))
    h_a.append(ROOT.TH1F(f'{name[f]}_a', f'{name[f]} all', 10, 0., 10.))
    h_sup.append(ROOT.TH1F(f'{name[f]}_s', f'{name[f]} sup', 10, 0., 10.))
    for entry in file[f].outtree:
        for i in range(len(entry.charge)):
            if entry.pad_charge[i*10] <= 0: continue
            mult_a = 1
            mult_sup = 1
            for pad in range(1, 10):
                if entry.pad_charge[i*10 + pad] > 0:
                    mult_a += 1
                    if abs(entry.pad_time[i*10 + pad]-entry.pad_time[i*10])>=4  or entry.pad_charge[i*10 + pad]/entry.pad_charge[i*10]>=0.08:
                        mult_sup += 1
            h_a[-1].Fill(mult_a)
            h_sup[-1].Fill(mult_sup)
    if h_a[-1].Integral() * h_sup[-1].Integral() == 0:
        continue
    h_a[-1].Scale(1./h_a[-1].Integral())
    h_sup[-1].Scale(1./h_sup[-1].Integral())
    h_a[-1].SetLineColor(color[f])
    h_sup[-1].SetLineColor(color[f])
    h_sup[-1].SetLineStyle(2)
    hs.Add(h_a[-1])
    hs.Add(h_sup[-1])
            
hs.Draw('hist nostack')
c.Draw()
c.BuildLegend()
```
